```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
# CSV
df = pd.read_csv('data/acc.csv', index_col=0)

# 
plt.figure(figsize=(10, 5))

# 
plt.plot(df.index, df['x'], label='acc_x')
plt.plot(df.index, df['y'], label='acc_y')
plt.plot(df.index, df['z'], label='acc_z')
# plt.plot(df.index, df['Absolute'], label='acc_Absolute')

# xy
plt.xlabel('time (s)')
plt.ylabel('acceleration (m/s^2)')

# 
plt.legend()

# 
plt.show()
```

```python
# CSV
df = pd.read_csv('data/acc.csv', index_col=0)

# 
df['x_filter'] = df['x'].rolling(100).mean()
df['y_filter'] = df['y'].rolling(100).mean()
df['z_filter'] = df['z'].rolling(100).mean()
# df['Absolute'] = df['Absolute'].rolling(100).mean()

# 
plt.figure(figsize=(10, 5))

# 
plt.plot(df.index, df['x_filter'], label='acc_x')
plt.plot(df.index, df['y_filter'], label='acc_y')
plt.plot(df.index, df['z_filter'], label='acc_z')
# plt.plot(df.index, df['Absolute'], label='acc_Absolute')

# 
# acc_z8time
df[df['x_up'] > 9].index
df[df['x_down'] > -9].index
df[df['y_up'] > 9].index
df[df['y_down'] > -9].index
df[df['z_up'] > 9].index
df[df['z_down'] > -9].index


# xy
plt.xlabel('time (s)')
plt.ylabel('acceleration (m/s^2)')

# 
plt.legend()

# 
plt.show()
```
